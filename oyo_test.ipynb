```python
#Import all 4 datasets provided in the previous question.

import pandas as pd;
TableA = pd.read_csv("TableA.csv",parse_dates=True)
TableB = pd.read_csv("TableB.csv" ,parse_dates=True)
TableC = pd.read_csv("TableC.csv",parse_dates=True)
TableD = pd.read_csv("TableD.csv")

# Merging all months data into one single dataframe
frames=[TableA,TableB,TableC]
final_data=pd.concat(frames)

# Adding new column month so that it is easier to group by
final_data['date'] = pd.to_datetime(final_data['date'])
final_data['month'] = final_data['date'].dt.strftime('%b')

final_data.head()
```
